```python
import plotly.express as px
```

```python
data = dataset.query("country == 'France' & province == 'Mayotte'")
fig = px.line(data, x="Date", y=["Confirmed","Recovered","Deaths"],markers=True,
              color_discrete_map={
                 "Confirmed": "#0072f0",
                 "Recovered": "#00b6cb",
                 "Deaths"   : "#f10096"
             },
              title="Covid Patients over time", template="plotly_white",height=600)

fig.update_layout(yaxis_range=[0,40000])
fig.show()
```
